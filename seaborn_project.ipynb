**CATEGORICAL**

```python
sns.catplot(x="AGE",y='EDUCATION',data=data,hue="EDUCATION",jitter=False)
```

**BOX PLOT**

```python
chart=sns.catplot(x="EDUCATION",y='AGE',hue='EDUCATION',data=data,kind="box")
chart.set_xticklabels(rotation=45,horizontalalignment='right',fontsize='x-large',fontweight='light',color='y' )
chart.set_yticklabels(fontsize='x-large',fontweight='light',color='y' )
```

**BAR PLOT**

**COUNT PLOT**

```python
ch=sns.catplot(x="MARITAL_STATUS",kind='count',hue='MARITAL_STATUS',data=data)
ch.set_xticklabels(rotation=45,horizontalalignment='right',fontsize='x-large',fontweight='light' )
```
